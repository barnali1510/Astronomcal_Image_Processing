### Calculating Accuracy with 100 Samples of each

```python
base_dir = "Astronomical_100_Dataset"

dataset_path = os.listdir(base_dir)
print (dataset_path)  #what kinds of classes are in this dataset
print("Types of classes labels found: ", len(dataset_path))

train = ImageDataGenerator(rescale = 1/255)
validation = ImageDataGenerator(rescale = 1/255)

import splitfolders
splitfolders.ratio(base_dir, output="output100", seed=1337, ratio=(.8, 0.1,0.1)) 

img_height,img_width=180,180
batch_size=28

train_dataset = train.flow_from_directory('output100/train',
                                         target_size=(img_height, img_width),
                                         batch_size=batch_size,
                                         class_mode='categorical')

validation_dataset = validation.flow_from_directory('output100/val',
                                         target_size=(img_height, img_width),
                                         batch_size=batch_size,
                                         class_mode='categorical')

train_dataset.class_indices
validation_dataset.class_indices
train_dataset.classes
validation_dataset.classes

model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(2,(3,3), activation = 'relu', input_shape = (180,180,3)),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(4,(3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(6,(3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    ##
                                    tf.keras.layers.Flatten(),
                                    ##
                                    tf.keras.layers.Dense(512, activation='relu'),
                                    ##
                                    tf.keras.layers.Dense(3, activation='softmax')
                                    ])

model.summary()

model.compile(optimizer=Adam(lr=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

earlystopping = callbacks.EarlyStopping(monitor ="val_loss", 
                                        mode ="min", patience = 5, 
                                        restore_best_weights = True)

model_fit = model.fit(train_dataset, validation_data=validation_dataset, epochs=10)

model_fit.history

# to plot the training and validation accuracy values
plt.plot(model_fit.history['accuracy'])
plt.plot(model_fit.history['val_accuracy'])
plt.grid()
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# to plot the training and loss accuracy values
plt.plot(model_fit.history['loss'])
plt.plot(model_fit.history['val_loss'])
plt.grid()
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

model_fit.history.keys()

model_fit.history['accuracy']
model_fit.history['val_accuracy']

from statistics import mean 

avg_accuracy = mean(model_fit.history['accuracy'])
print("Average Traing Accuracy = ") 
print(avg_accuracy)

avg_val_accuracy = mean(model_fit.history['val_accuracy'])
print("Average Test Accuracy = ")
print(avg_val_accuracy)

Y_pred = model.predict(train_dataset)
y_pred = np.argmax(Y_pred, axis=1)

print(classification_report(train_dataset.classes, y_pred))

import itertools
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
def plot_confusion_matrix(cm, classes,
                         normalize=False,
                         title='Confusion_Matrix',
                         cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized Confusion Matrix")
    else:
        print("Confusion Matrix withour Normalization")
        
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, weight='bold', fontsize=16)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=16)
    plt.yticks(tick_marks, classes, fontsize=16)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center", fontsize=12, weight='bold',
                 color="red" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('Actual Class', fontsize=16, weight='bold')
    plt.xlabel('Predicted Class', fontsize=16, weight='bold')
    
#compute Confusion Matrix
cnf_matrix = confusion_matrix(train_dataset.classes, y_pred)
np.set_printoptions(precision=2)

#plot normalized Confusion Matrix
plt.figure(figsize=(10, 10))
plot_confusion_matrix(cnf_matrix, classes = ['Backhole', 'Galaxy', 'Nebula'], normalize=True, title='Normalized Confusion Matrix')
plt.show()
```

```python
A1 = avg_accuracy
B1 = avg_val_accuracy
print(A1)
print(B1)
```

```python
base_dir = "Astronomical_500_Dataset"

dataset_path = os.listdir(base_dir)
print (dataset_path)  #what kinds of classes are in this dataset
print("Types of classes labels found: ", len(dataset_path))

train = ImageDataGenerator(rescale = 1/255)
validation = ImageDataGenerator(rescale = 1/255)

import splitfolders
splitfolders.ratio(base_dir, output="output500", seed=1337, ratio=(.8, 0.1,0.1)) 

img_height,img_width=180,180
batch_size=28

train_dataset = train.flow_from_directory('output500/train',
                                         target_size=(img_height, img_width),
                                         batch_size=batch_size,
                                         class_mode='categorical')

validation_dataset = validation.flow_from_directory('output500/val',
                                         target_size=(img_height, img_width),
                                         batch_size=batch_size,
                                         class_mode='categorical')

train_dataset.class_indices
validation_dataset.class_indices
train_dataset.classes
validation_dataset.classes

model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(2,(3,3), activation = 'relu', input_shape = (180,180,3)),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(4,(3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(6,(3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    ##
                                    tf.keras.layers.Flatten(),
                                    ##
                                    tf.keras.layers.Dense(512, activation='relu'),
                                    ##
                                    tf.keras.layers.Dense(3, activation='softmax')
                                    ])

model.summary()

model.compile(optimizer=Adam(lr=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

earlystopping = callbacks.EarlyStopping(monitor ="val_loss", 
                                        mode ="min", patience = 5, 
                                        restore_best_weights = True)

model_fit = model.fit(train_dataset, validation_data=validation_dataset, epochs=10)

model_fit.history

# to plot the training and validation accuracy values
plt.plot(model_fit.history['accuracy'])
plt.plot(model_fit.history['val_accuracy'])
plt.grid()
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# to plot the training and loss accuracy values
plt.plot(model_fit.history['loss'])
plt.plot(model_fit.history['val_loss'])
plt.grid()
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

model_fit.history.keys()

model_fit.history['accuracy']
model_fit.history['val_accuracy']

from statistics import mean 

avg_accuracy = mean(model_fit.history['accuracy'])
print("Average Traing Accuracy = ") 
print(avg_accuracy)

avg_val_accuracy = mean(model_fit.history['val_accuracy'])
print("Average Test Accuracy = ")
print(avg_val_accuracy)

Y_pred = model.predict(train_dataset)
y_pred = np.argmax(Y_pred, axis=1)

print(classification_report(train_dataset.classes, y_pred))

import itertools
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
def plot_confusion_matrix(cm, classes,
                         normalize=False,
                         title='Confusion_Matrix',
                         cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized Confusion Matrix")
    else:
        print("Confusion Matrix withour Normalization")
        
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, weight='bold', fontsize=16)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=16)
    plt.yticks(tick_marks, classes, fontsize=16)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center", fontsize=12, weight='bold',
                 color="red" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('Actual Class', fontsize=16, weight='bold')
    plt.xlabel('Predicted Class', fontsize=16, weight='bold')
    
#compute Confusion Matrix
cnf_matrix = confusion_matrix(train_dataset.classes, y_pred)
np.set_printoptions(precision=2)

#plot normalized Confusion Matrix
plt.figure(figsize=(10, 10))
plot_confusion_matrix(cnf_matrix, classes = ['Backhole', 'Galaxy', 'Nebula'], normalize=True, title='Normalized Confusion Matrix')
plt.show()
```
